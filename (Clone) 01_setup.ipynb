```python
%sql

CREATE CATALOG IF NOT EXISTS main;
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS main.healthcare_project_pipe;
```

```python
%sql
DROP TABLE IF EXISTS main.healthcare_project_pipe.raw_members;

CREATE TABLE main.healthcare_project_pipe.raw_members
AS
SELECT 
    CAST(member_id AS INT) AS member_id,
    member_name,
    gender,
    city,
    state,
    CAST(join_date AS DATE) AS join_date
FROM read_files(
    '/Volumes/main/healthcare_project_pipe/raw_data/members/',
    format => 'csv',
    header => true
);
```

```python
%sql
DROP TABLE IF EXISTS main.healthcare_project_pipe.raw_claims;

CREATE TABLE main.healthcare_project_pipe.raw_claims
AS
SELECT 
    CAST(claim_id AS INT) AS claim_id,
    CAST(member_id AS INT) AS member_id,
    CAST(provider_id AS INT) AS provider_id,
    CAST(claim_amount AS DOUBLE) AS claim_amount,
    claim_status,
    TRY_CAST(claim_date AS DATE) AS claim_date
FROM read_files(
    '/Volumes/main/healthcare_project_pipe/raw_data/claims/',
    format => 'csv',
    header => true
);
```

```python
%sql
DROP TABLE IF EXISTS main.healthcare_project_pipe.raw_providers;

CREATE TABLE main.healthcare_project_pipe.raw_providers
AS
SELECT 
    CAST(provider_id AS INT) AS provider_id,
    provider_name,
    specialty,
    city
FROM read_files(
    '/Volumes/main/healthcare_project_pipe/raw_data/providers/',
    format => 'csv',
    header => true
);
```

```python
%sql
DROP TABLE IF EXISTS main.healthcare_project_pipe.raw_pharmacy;

CREATE TABLE main.healthcare_project_pipe.raw_pharmacy
AS
SELECT 
    CAST(member_id AS INT) AS member_id,
    CAST(rx_id AS INT) AS rx_id,
    drug_name,
    CAST(rx_cost AS DOUBLE) AS rx_cost,
    TRY_CAST(rx_date AS DATE) AS rx_date
FROM read_files(
    '/Volumes/main/healthcare_project_pipe/raw_data/pharmacy/',
    format => 'csv',
    header => true
);
```
