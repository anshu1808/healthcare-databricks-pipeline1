```python
from pyspark.sql.functions import current_timestamp

df_claims_raw = spark.table("main.healthcare_project_pipe.raw_claims")
df_members_raw = spark.table("main.healthcare_project_pipe.raw_members")
df_pharmacy_raw = spark.table("main.healthcare_project_pipe.raw_pharmacy")
df_providers_raw = spark.table("main.healthcare_project_pipe.raw_providers")

display(df_members_raw)
display(df_claims_raw)
display(df_pharmacy_raw)
display(df_providers_raw)
```

```python
from pyspark.sql.functions import current_timestamp

df_members_bronze = df_members_raw \
    .withColumn("ingestion_time", current_timestamp())

df_claims_bronze = df_claims_raw \
    .withColumn("ingestion_time", current_timestamp())

df_provider_bronze = df_providers_raw \
    .withColumn("ingestion_time", current_timestamp())

df_pharmacy_bronze = df_pharmacy_raw \
    .withColumn("ingestion_time", current_timestamp())
```

```python
df_members_bronze.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("main.healthcare_project_pipe.bronze_members")

df_claims_bronze.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("main.healthcare_project_pipe.bronze_claims")

df_provider_bronze.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("main.healthcare_project_pipe.bronze_providers")

df_pharmacy_bronze.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("main.healthcare_project_pipe.bronze_pharmacy")
```
