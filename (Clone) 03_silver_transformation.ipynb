```python
from pyspark.sql.functions import current_timestamp

df_claims_silver = spark.table("main.healthcare_project_pipe.bronze_claims")
df_members_silver = spark.table("main.healthcare_project_pipe.bronze_members")
df_pharmacy_silver = spark.table("main.healthcare_project_pipe.raw_pharmacy")
df_providers_silver = spark.table("main.healthcare_project_pipe.bronze_providers")

display(df_claims_silver)
display(df_members_silver)
display(df_pharmacy_silver)
display(df_providers_silver)
```

```python
df_claims_silver = df_claims_silver.dropDuplicates()
df_members_silver = df_members_silver.dropDuplicates()
df_pharmacy_silver = df_pharmacy_silver.dropDuplicates()
df_providers_silver = df_providers_silver.dropDuplicates()
```

```python
from pyspark.sql.functions import col
from pyspark.sql.functions import *

df_claims = df_claims_silver.filter(
    col("claim_id").isNotNull()
)

df_claims = df_claims.filter(
    col("claim_amount") > 0
)
```

```python
df_claims = df_claims.withColumn(
    "claim_status",
    upper(trim(col("claim_status")))
)

df_claims = df_claims.withColumn(
    "claim_year",
    year("claim_date")
).withColumn(
    "claim_month",
    month("claim_date")
)

display(df_claims)
```

```python
df_claims.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("main.healthcare_project_pipe.silver_claims")
```

```python
df_members_silver.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("main.healthcare_project_pipe.silver_members")

df_providers_silver.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("main.healthcare_project_pipe.silver_providers")

df_pharmacy_silver.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("main.healthcare_project_pipe.silver_pharmacy")
```
